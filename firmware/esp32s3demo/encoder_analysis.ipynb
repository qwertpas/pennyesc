```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('dyno_logs/dyno_log_20260123_115109.csv')
df
```

```python
plt.figure(figsize=(8,8))
plt.scatter(df['v_angle']%2*np.pi, df['p_position']%2*np.pi, s=2, alpha=0.8, c=df['v_vel'])

plt.xlim(0, 2*np.pi)
plt.ylim(0, 2*np.pi)
plt.xlabel("vertiq angle")
plt.ylabel("penny angle")
plt.title("yellow is faster, purple is slower")
plt.gca().set_aspect('equal')
```
